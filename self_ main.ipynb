```python
with open('sample.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize by words
words = text.split()

# Print the tokenized words
print(words)
```

```python
len(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
stop = stopwords.words("english")
```

```python
stop_words = list(punctuation)  + stop
```

```python
cleaned_txt = []
for i in words:
    if i not in stop_words:
        cleaned_txt.append(i)
print(cleaned_txt)
```

```python
import re 
```

```python
cleaned_txt = " ".join(cleaned_txt)
cleaned_txt
```

```python
cleaned_txt = re.sub('[0-9]' , "" , cleaned_txt) # digit removal from string
cleaned_txt = re.sub(r'\[\d+\]' , "" , cleaned_txt)  # digit removal within brackets
```

```python
cleaned_txt
```

```python
len(cleaned_txt)
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize
```

```python
word_token = word_tokenize(cleaned_txt)
word_token
```

```python
from nltk.stem import PorterStemmer
```

```python
p = PorterStemmer()
porter_stemmer = [p.stem(word) for word in word_token]
porter_stemmer
```

```python
cleaned_txt = " ".join(porter_stemmer)
cleaned_txt
```

```python
sent_token = sent_tokenize(cleaned_txt)
sent_token
```

```python
from textblob import TextBlob

sentence_list = []
polarity_act = []
polarity_list =[]

for i in sent_token:
    sentence = i
    blob = TextBlob(sentence)  # Create a TextBlob object
    polarity = blob.sentiment.polarity  # Get the polarity score 
    sentence_list.append(sentence)
    polarity_act.append(polarity)
    if polarity>0:
        print(f"Sentence: {sentence}")
        print(f"Polarity: {polarity}")
        print(f"Positive:{1}")
        polarity_list.append(1)
        print("\t")
            
    elif polarity==0:
        print(f"Sentence: {sentence}")
        print(f"Polarity: {polarity}")
        print(f"Neutral :{0}")
        polarity_list.append(0)
        print("\t")
    else:
        print(f"Sentence: {sentence}")
        print(f"Polarity: {polarity}")
        print(f"Negative: {-1}")
        polarity_list.append(-1)
        print("\t")
```
