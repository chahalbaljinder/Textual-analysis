```python
with open('sample.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize by words
words = text.split()
words = [word.lower() for word in words]
# Print the tokenized words
print(words)
```

```python
len(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
stop = stopwords.words("english")
```

```python
stop_words = list(punctuation)  + stop
```

```python
cleaned_txt = []
for i in words:
    if i not in stop_words:
        cleaned_txt.append(i)
print(cleaned_txt)
```

```python
import re 
```

```python
cleaned_txt = " ".join(cleaned_txt)
cleaned_txt
```

```python
cleaned_txt = re.sub('[0-9]' , "" , cleaned_txt) # digit removal from string
cleaned_txt = re.sub(r'\[\d+\]' , "" , cleaned_txt)  # digit removal within brackets
```

```python
cleaned_txt
```

```python
len(cleaned_txt)
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize
```

```python
word_token = word_tokenize(cleaned_txt)
word_token
```

```python
from nltk.stem import PorterStemmer
```

```python
p = PorterStemmer()
porter_stemmer = [p.stem(word) for word in word_token]
porter_stemmer
```

```python
cleaned_txt = " ".join(porter_stemmer)
cleaned_txt
```

```python
sent_token = sent_tokenize(cleaned_txt)
sent_token
```

```python
from textblob import TextBlob

sentence_list = []
polarity_act = []
polarity_list =[]

for i in sent_token:
    sentence = i
    blob = TextBlob(sentence)  # Create a TextBlob object
    polarity = blob.sentiment.polarity  # Get the polarity score 
    sentence_list.append(sentence)
    polarity_act.append(polarity)
    if polarity>0:
        print(f"Sentence: {sentence}")
        print(f"Polarity: {polarity}")
        print(f"Positive:{1}")
        polarity_list.append(1)
        print("\t")
            
    elif polarity==0:
        print(f"Sentence: {sentence}")
        print(f"Polarity: {polarity}")
        print(f"Neutral :{0}")
        polarity_list.append(0)
        print("\t")
    else:
        print(f"Sentence: {sentence}")
        print(f"Polarity: {polarity}")
        print(f"Negative: {-1}")
        polarity_list.append(-1)
        print("\t")
```

```python
print(len(sentence_list))
print(len(polarity_act))
print(len(polarity_list))
```

```python
polar_data = pd.DataFrame({
    "Sentence" : sentence_list, 
    "Polarity_val" : polarity_act,
    "Polarity_score" : polarity_list
})
```

```python
polar_data.to_csv("polar_data.csv")
```

```python
dataset = pd.read_csv("polar_data.csv")
```

```python
dataset.drop(dataset.columns[0] , axis=1 ,inplace=True)
```

```python
dataset.head(3)
```

```python
x=dataset[["Polarity_val"]]
y=dataset["Polarity_score"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(criterion='squared_error', splitter='random')
dt.fit(x_train , y_train)
```

```python
dt.score(x_train , y_train)*100  , dt.score(x_test , y_test)*100
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
pg = {
              "criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"],
              "splitter": ["best", "random"]             
             }
```

```python
gd = GridSearchCV(DecisionTreeRegressor() , param_grid= pg)
```

```python
gd.fit(x_train , y_train)
```

```python
gd.score(x_train , y_train)*100 , gd.score(x_test , y_test)*100
```
